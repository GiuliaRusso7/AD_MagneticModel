```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os
```

# get in dipoles folder

```python
ChosenMomentumBeam = "0.3" # 3.574
```

```python
dipoles_directory_path = "/eos/home-r/russogi/WINDOWS/Desktop/AD_ELENA_scripts_data/AD_MagneticModel/dipoles/"

dipoles_families = os.listdir(dipoles_directory_path)

temp = os.listdir(dipoles_directory_path)
FilesForChosenMomentum = []
for file in temp:
    if "%s"%ChosenMomentumBeam in file:
        FilesForChosenMomentum.append(file)
        
FitFiles, StrenghtFiles, dipoles_families = [], [], []

for file in FilesForChosenMomentum:
    if "FIT" in file:
        FitFiles.append(file)
    elif "Bending" in file:
        StrenghtFiles.append(file)
    elif "StrenghtsDipoles" in file:
        pass
    else:
        dipoles_families.append(file)
```

```python
dipoles_families = ["BHN", "BHS", "BHW"]
```

# compare measures with fit

```python
def LinearLaw (x, a, b):
    y = a*x + b
    return y
```

```python
os.chdir(dipoles_directory_path)

f, ax_d = plt.subplots(1, len(dipoles_families), figsize = (35,6), sharex = True, sharey = False)

for i_dip, dipole in enumerate(dipoles_families):
    
    df_strenght = pd.read_csv("%s"%StrenghtFiles[i_dip])
    df_fit = pd.read_csv("%s"%FitFiles[i_dip])
    
    ax_d[i_dip].plot(df_strenght.Current, df_strenght.Calculated, "x-g")
    ax_d[i_dip].plot(df_strenght.Current, df_strenght.Measured, "o-g")
    
    ax_d[i_dip].plot(df_strenght.Current, LinearLaw(df_strenght.Current, *df_fit["linear"].values), "--k", lw = 2)

    ax_d[i_dip].set_xlabel ("Current [A]", size = 15)
    ax_d[i_dip].tick_params (axis = "both", which = "major", labelsize = 13)
    ax_d[i_dip].set_title("%s"%dipole, size = 15)
ax_d[0].set_ylabel (r"$\theta$ [rad]", size = 15)
```
